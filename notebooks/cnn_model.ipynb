```python
import sys
import os
sys.path.append(os.path.abspath(".."))

# Import Keras libraries
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Activation
from tensorflow.keras.optimizers import SGD
from tensorflow.keras.layers import Dense, Dropout,BatchNormalization
from tensorflow.keras.layers import Conv2D, MaxPooling2D
from tensorflow.keras import backend as K
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.preprocessing import image
from tensorflow.keras import layers
from tensorflow.keras.applications.efficientnet import preprocess_input


# Libraries voor Early stopping, ModelCheckpoint en load_model
from tensorflow.keras.models import load_model
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.callbacks import ModelCheckpoint
from tensorflow.keras.callbacks import ReduceLROnPlateau
import os
import pandas as pd
import numpy as np

from ml_models.cnn_model import build_model, CNNModelRunner
from utils.plots import plot_runner_method
```

```python
# Read in image data
train_image_folder_path = os.path.join(os.getcwd(), 'data', 'raw_images', 'archive', 'images', 'train')
val_image_folder_path = os.path.join(os.getcwd(), 'data', 'raw_images', 'archive', 'images', 'val')

train_label_folder_path = os.path.join(os.getcwd(), 'data', 'raw_images', 'archive', 'labels', 'train', 'labels.csv')
val_label_folder_path = os.path.join(os.getcwd(), 'data', 'raw_images', 'archive', 'labels', 'val', 'labels.csv')

df_train_label = pd.read_csv(train_label_folder_path)

train_label_dict = {
    row['image_filename']: row[1:].values.astype(np.int32)
    for _, row in df_train_label.iterrows()
}

df_train_label

df_val_label = pd.read_csv(val_label_folder_path)

val_label_dict = {
    row['image_filename']: row[1:].values.astype(np.int32)
    for _, row in df_val_label.iterrows()
}

```

```python
train_image_file_paths = []
y_train_dict = {
    "head": [],
    "neck": [],
    "shoulder": [],
    "thoraric": [],
    "lumbar": [],
}

for filename in os.listdir(train_image_folder_path):
    if filename.lower().endswith(('.jpg', '.png', '.jpeg')):
        if filename in df_train_label['image_filename'].values:
            train_image_file_paths.append(os.path.join(train_image_folder_path, filename))
            labels = train_label_dict[filename]
            y_train_dict["head"].append(labels[0])
            y_train_dict["neck"].append(labels[1])
            y_train_dict["shoulder"].append(labels[2])
            y_train_dict["thoraric"].append(labels[3])
            y_train_dict["lumbar"].append(labels[4])
            
for key in y_train_dict:
    y_train_dict[key] = np.array(y_train_dict[key])


val_image_file_paths = []
y_val_dict = {
    "head": [],
    "neck": [],
    "shoulder": [],
    "thoraric": [],
    "lumbar": [],
}

for filename in os.listdir(val_image_folder_path):
    if filename.lower().endswith(('.jpg', '.png', '.jpeg')):
        if filename in df_val_label['image_filename'].values: #.values because image_filename not index
            val_image_file_paths.append(os.path.join(val_image_folder_path, filename))
            labels = val_label_dict[filename]
            y_val_dict["head"].append(labels[0])
            y_val_dict["neck"].append(labels[1])
            y_val_dict["shoulder"].append(labels[2])
            y_val_dict["thoraric"].append(labels[3])
            y_val_dict["lumbar"].append(labels[4])
            
for key in y_val_dict:
    y_val_dict[key] = np.array(y_val_dict[key])
```

```python
y_train_dict
```

```python
#preprocessing
def preprocess_image(path):
    image = tf.io.read_file(path)
    image = tf.image.decode_image(image, channels=3) # Convert raw image bytes to tensor
    image.set_shape([None, None, 3])
    image = tf.image.resize(image, [224, 224])
    image = preprocess_input(image)  # This step replaces manual normalization
    return image
```

```python
def load_data(image_path, label_dict):
    image = preprocess_image(image_path)
    return image, label_dict
```

```python
BATCH_SIZE = 32
```

```python
#Create tensorflow dataset objects
train_ds = tf.data.Dataset.from_tensor_slices((train_image_file_paths, y_train_dict))
train_ds = train_ds.map(load_data, num_parallel_calls=tf.data.AUTOTUNE)
train_ds = train_ds.shuffle(1024).batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)

val_ds = tf.data.Dataset.from_tensor_slices((val_image_file_paths, y_val_dict))
val_ds = val_ds.map(load_data, num_parallel_calls=tf.data.AUTOTUNE)
val_ds = val_ds.batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)
```

```python
train_ds
```

```python
config = {
    # Training hyperparameters
    "learning_rate": 0.0001,
    "optimizer": "rmsprop",
    "epochs": 10,
    "batch_size": 32,

    # Model architecture
    "dropout_rate": 0.5,
    "dense_units": 256,

    # Loss weights
    "loss_weights": {
        "head": 1.0,
        "neck": 1.0,
        "shoulder": 1.0,
        "thoraric": 1.0,
        "lumbar": 1.0
    },

    # Loss functions
    "losses": {
        "head": "binary_crossentropy",
        "neck": "sparse_categorical_crossentropy",
        "shoulder": "binary_crossentropy",
        "thoraric": "binary_crossentropy",
        "lumbar": "sparse_categorical_crossentropy"
    },

    # Metrics per output
    "metrics": {
        "head": "accuracy",
        "neck": "accuracy",
        "shoulder": "accuracy",
        "thoraric": "accuracy",
        "lumbar": "accuracy"
    },

    # Callbacks
    "callbacks": {
        "monitor": "val_loss",
        "early_stopping_patience": 5,
        "reduce_lr_patience": 3,
        "reduce_lr_factor": 0.2,
        "min_lr": 0.000001,
        "save_best_only": True,
        "checkpoint_path": "efficientnet_multioutput.keras"
    }
}
```

```python
# model
# Transfer learning
# Instantiatie van de convolutionele basis: EfficientNet B0
conv_base = keras.applications.EfficientNetB0(
    include_top=False, # We use our own classifier, not the 1000 classes of imagenet
    weights="imagenet",
    input_shape=(224, 224, 3), 
    pooling=None,
    classifier_activation="softmax",
    name="efficientnetb0",
)
```

```python
model = build_model(conv_base, config)

model.compile(
    optimizer=config["optimizer"],
    loss=config["losses"],
    metrics=config["metrics"],
    loss_weights=config["loss_weights"]
)

callbacks = [
    keras.callbacks.ModelCheckpoint(
        filepath=config["callbacks"]["checkpoint_path"],
        save_best_only=config["callbacks"]["save_best_only"],
        monitor=config["callbacks"]["monitor"]
    ),
    keras.callbacks.ReduceLROnPlateau(
        monitor=config["callbacks"]["monitor"],
        factor=config["callbacks"]["reduce_lr_factor"],
        patience=config["callbacks"]["reduce_lr_patience"],
        min_lr=config["callbacks"]["min_lr"]
    ),
    keras.callbacks.EarlyStopping(
        monitor=config["callbacks"]["monitor"],
        patience=config["callbacks"]["early_stopping_patience"],
        restore_best_weights=True
    )
]
```

```python
history = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=config[epochs],
    callbacks=callbacks
)
```

```python
runner = CNNModelRunner(model=model, val_ds=val_ds, output_names=list(config["metrics"].keys()), config=config)
metrics = runner.evaluate()
runner.log_to_mlflow(experiment_name="CNN_Model")
```
