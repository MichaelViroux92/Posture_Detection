```python
file_path_labels_train = os.path.join(os.getcwd(), 'data', 'raw_images', 'archive', 'labels', 'train', 'labels.csv')
file_path_labels_val = os.path.join(os.getcwd(), 'data', 'raw_images', 'archive', 'labels', 'val', 'labels.csv')

df_train_label = pd.read_csv(file_path_labels_train)
df_val_label = pd.read_csv(file_path_labels_val)

df_train_label.head()
```

```python
df_val_label.head()
```

```python
df_train_label['image_filename'] = df_train_label['image_filename'].astype(str) + '_train'
df_train_label.head()
```

```python
df_val_label['image_filename'] = df_val_label['image_filename'].astype(str) + '_val'
df_val_label.head()
```

```python
df_label = pd.concat([df_train_label, df_val_label])
df_label.head()
```

```python
df_label.tail()
```

```python
# Check class imbalances 
fig, axes = plt.subplots(1, 5, figsize=(12, 4)) 

# Head
head_counts = df_label['Head'].value_counts()
axes[0].bar(head_counts.index, head_counts.values, color=['orange', 'blue'])
axes[0].set_title("Head")
axes[0].set_xticks(head_counts.index)

# Neck
neck_counts = df_label['Neck'].value_counts()
axes[1].bar(neck_counts.index, neck_counts.values, color=['orange', 'blue', 'green'])
axes[1].set_title("Neck")
axes[1].set_xticks(neck_counts.index)


# Shoulder
shoulder_counts = df_label['Shoulder'].value_counts()
axes[2].bar(shoulder_counts.index, shoulder_counts.values, color=['orange', 'blue'])
axes[2].set_title("Shoulder")
axes[2].set_xticks(shoulder_counts.index)

# Thoraric Spine
thoraric_spine_counts = df_label['Thoraric_Spine'].value_counts()
axes[3].bar(thoraric_spine_counts.index, thoraric_spine_counts.values, color=['orange', 'blue'])
axes[3].set_title("Thoraric Spine")
axes[3].set_xticks(thoraric_spine_counts.index)

# Lumbar Spine
lumbar_spine_counts = df_label['Lumbar_Spine'].value_counts()
axes[4].bar(lumbar_spine_counts.index, lumbar_spine_counts.values, color=['orange', 'blue', 'green'])
axes[4].set_title("Lumbar Spine")
axes[4].set_xticks(lumbar_spine_counts.index)


plt.tight_layout()
plt.show()
```
